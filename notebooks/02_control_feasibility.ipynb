# Control feasibility

This section describes candidate Eurostat control variables. Each row uses an exact dataset code and filter.

```python
from pathlib import Path

import pandas as pd

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
feasibility_path = PROJECT_ROOT / 'outputs' / 'control_feasibility.csv'
controls_path = PROJECT_ROOT / 'data' / 'processed' / 'verified_control_candidates.csv'

feasibility = pd.read_csv(feasibility_path)
controls = pd.read_csv(controls_path)
feasibility
```

```python
# keep only columns approved by the feasibility audit
approved = feasibility.loc[feasibility['merge_approved'].eq('yes'), 'variable_name'].tolist()
approved
```

```python
missingness = controls[['geo', 'year'] + approved].copy()
missingness = missingness.groupby('year')[approved].apply(lambda frame: frame.notna().sum()).reset_index()
missingness
```

```python
feasibility.to_csv(PROJECT_ROOT / 'outputs' / 'control_feasibility_notebook.csv', index=False)
feasibility.to_markdown(PROJECT_ROOT / 'outputs' / 'control_feasibility_notebook.md', index=False)
missingness.to_csv(PROJECT_ROOT / 'outputs' / 'control_missingness_by_year.csv', index=False)
```

The audit records source feasibility and missingness. It does not estimate any model.
